```python
print('1. checking the presence of outlier in "bmi" ->')
x_bmi = insuranceData['bmi']
print('   1st quartile:',np.quantile(x_bmi,q=0.25))
print('   3rd quartile:',np.quantile(x_bmi,q=0.75))
interquartile_range=np.quantile(x_bmi,q=0.75)-np.quantile(x_bmi,q=0.25)
print('   interquartile range=>3rd quartile - 1st quartile=',interquartile_range)
range = interquartile_range*1.5
print('   range:',range)
max_q = interquartile_range*1.5 + np.quantile(x_bmi,q=0.75)
min_q = interquartile_range*1.5 - np.quantile(x_bmi,q=0.25)
print('max_q:',max_q)
print('min_q:',min_q)
print('   checking if any point is outlier->')
outlier = []
for x in x_bmi:
    if(x>max_q):
       print('point ',x,' is an outliar above the interquartile range')
       outlier.append(x)
    elif(x<min_q):
       print('point ',x,' is an outliar below the interquartile range')
       outlier.append(x)
print('   total number of outlier is: ',len(outlier))
print('====================')
print('drawing box plot:')
plt.boxplot(insuranceData['bmi'])
```
