# Preprocess rows of the “text” column
a. Remove unwanted characters

b. Convert text to lowercase

c. Remove unwanted spaces

d. Remove stopwords

```python
def remove_punctuation(text):
  no_pun="".join([p for p in text if p not in string.punctuation])
  return no_pun
```

```python
df['text']=df['text'].apply(lambda x: remove_punctuation(x))
df['text'].head(5)
```

```python
df['text']=df['text'].apply(lambda x: x.lower())
```

```python
df['text'].head(5)
```

```python
df['text']=df['text'].apply(lambda x: " ".join(x.split()))
```

```python
df.head(5)
```

```python
stop = stopwords.words('english')
```

```python
df['text'] = df['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
df['text'].head(5)
```

```python
df['labels']=df['gender'].astype(str)+', '+df['age'].astype(str)+', '+df['topic'].astype(str)+', '+df['sign'].astype(str)
```

```python
df.head(10)
```

```python
df_new = df.loc[:, ['text','labels']]
```

```python
df_new
```

```python
x=df_new['text']
y=df_new['labels']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
# use CountVectorizer to create document-term matrices from X_train and X_test
vect = CountVectorizer(ngram_range=(1, 2))
x_train_dtm = vect.fit_transform(x_train)
x_test_dtm = vect.transform(x_test)
```

```python
x_train_dtm.shape
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
```

```python
y_train_trans = mlb.fit_transform(y_train)
y_test_trans = mlb.transform(y_test)
```

```python
y_train_trans
```

```python
import numpy as np
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
clf = OneVsRestClassifier(LogisticRegression(solver='lbfgs')).fit(x_train_dtm, y_train_trans)
```

```python
clf.score(x_test_dtm, y_test_trans)
```

```python
clf.score(x_train_dtm, y_train_trans)
```

```python
from sklearn.metrics import classification_report
predicted = clf.predict(x_test_dtm)
report = classification_report(y_test_trans, predicted)
print(report)
```
